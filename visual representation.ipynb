```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
file_path = "SupplyChainEmissionFactorsforUSIndustriesCommodities(2015_Summary_Industry).csv"
df = pd.read_csv(file_path)
co2e_df = df[df['Substance'].isin(['carbon dioxide', 'other GHGs'])]
top_industries = ( co2e_df.groupby('Industry Name')['Supply Chain Emission Factors with Margins']
                  .sum()
                  .sort_values(ascending=False)
                  head(20)
                 )
```

```python
plt.figure(figsize=(12,6))
top_industries.plot(kind='bar',color='orange',edgecolor='black')
plt.title('Top 10 Emitting Industries by CO₂e Supply Chain Emission Factors',fontsize=14)
plt.ylabel('emission factor' , fontsize = 10)
plt.xlabel('industry', fontsize = 10)
plt.show()
```

```python
top_industries = (co2e_df.groupby('Industry Name')['Supply Chain Emission Factors with Margins']
                  .sum()
                  .sort_values(ascending=False)
                  .head(10)
                 )
```
