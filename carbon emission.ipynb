```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
file_path = "SupplyChainEmissionFactorsforUSIndustriesCommodities(2015_Summary_Industry).csv"
df = pd.read_csv(file_path)
```

```python
df['Supply Chain Emission Factors with Margins'] = pd.to_numeric(
    df['Supply Chain Emission Factors with Margins'], errors='coerce'
)
```

```python
co2e_df = df[df['Substance'].isin(['carbon dioxide', 'other GHGs'])]
```

```python
top_industries = (
    co2e_df.groupby('Industry Name')['Supply Chain Emission Factors with Margins']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(12, 6))
top_industries.plot(kind='bar', color='tomato')
plt.title('Top 10 Emitting Industries by COe Supply Chain Emission Factors')
plt.ylabel('Emission Factor (kg COe per USD)')
plt.xlabel('Industry')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
